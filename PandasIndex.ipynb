```python
grades_dict = {'A':4, 'B':3.5, 'C':3, 'D':2.5}
```

```python
puan = pd.Series(puans_dict)
```

```python
dFrame = pd.DataFrame({'Marks':puan, 'Grades':grades})
```

```python
dFrame.iloc[2,:]
```
