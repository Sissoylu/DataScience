```python
A = np.round(10*np.random.rand(2,3)) #rand(2,3) => 3 elemanlı 2 tane array oluşturdu
```

```python
A
```

```python
A+3 #tüm elemenların 3 ile toplanmasını sağladık
```

```python
A+(np.arange(2).reshape(2,1))
```

```python
A
```

```python
B = np.round(10*np.random.rand(2,2)) 
```

```python
B
```

```python
C = np.hstack((A,B)) 
#must be tuple inside hstack
#hstack iki arrayin yatay olarak birleşmesini sağlar
#hstack için sütun satır eşitsizliği önemli değildir.
```

```python
C
```

```python
D = np.round(10*np.random.rand(2,2))
```

```python
E = np.vstack((B,D)) 
#eğer arraylerden biri 2x3 diğeri 2x2 ise vstack çalışmaz aynı sütun sayıısına sahip olmalı 
#arrayler 2x3 3x3 şeklindeyse vstack çalışır. Burada önemli olan sütün sayısının eşit olmasıdır.
```

```python
E
```

```python
F = np.round(10*np.random.rand(3,3))
```

```python
G = np.vstack((A,F))
```

```python
G 
```

```python
H = np.hstack((D,F))
```

```python
H
```

```python
shuf = np.random.permutation(np.arange(10)) #10 elemanlı bir array oluşturuldu.
```

```python
shuf
```

```python
shuf.sort() #sort() fonksiyonu arrayi küçükten büyüğe sıralamamızı sağlar.
```

```python
shuf
```

```python
#we can also use sort func like that;
np.sort(shuf)
```

```python
shuf = shuf[::-1] #en sondaki elemandan başlayarak ilk elemana kadar gider bu sayede arrayi tersten sıralamış olur
```

```python
shuf
```
